```python
df['begin_area'] = df['begin_area'].str.replace(',','')
```

```python
for label in ['area_type','begin_area_type']:
    
    for atype in df[label].dropna().unique():
        
        mask = df[label]==atype
        
        if label == 'area_type':
            df[atype[:20].lower().replace(' ','_')] = df[mask]['area']
        else:
            df['begin_'+atype[:20].lower().replace(' ','_')] = df[mask]['begin_area']
```
